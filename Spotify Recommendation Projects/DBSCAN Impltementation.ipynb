```python
import json
import pandas as pd
import seaborn as sns
import requests
from pathlib import Path
import os
```

```python
directory_path = 'C:\\Users\\lochoahernandez\\OneDrive - Cherry Creek School District\\Documents\\GitHub\\Machine-Learning\\Spotify Recommendation Projects\\Data Storage'

combined_data = []

for filename in os.listdir(directory_path):
        file_path = os.path.join(directory_path, filename)
        with open(file_path, 'r') as f:
            data = json.load(f)
            if isinstance(data, list):
                combined_data.extend(data)
            else:
                combined_data.append(data)

with open('combined.json', 'w') as f:
    json.dump(combined_data, f, indent=4)

print("Combined JSON file created with", len(combined_data), "items.")
```

```python
with open('combined.json', 'r') as f:
    combined_data = json.load(f)

all_playlists = []

if isinstance(combined_data, list):
    for entry in combined_data:
        if isinstance(entry, dict) and 'playlists' in entry:
            if isinstance(entry['playlists'], list):
                all_playlists.extend(entry['playlists'])
            else:
                all_playlists.append(entry['playlists'])
else:
    all_playlists = combined_data.get('playlists', [])

print("Total playlists found:", len(all_playlists))

```

```python
data = pd.DataFrame(all_playlists)
data
```

```python
all_tracks = []
for playlist in all_playlists:
    tracks = playlist.get("tracks", [])
    for track in tracks:
        all_tracks.append(track)

print("Total number of tracks found across all playlists:", len(all_tracks))
```
