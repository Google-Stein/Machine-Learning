```python
import json
import pandas as pd
import seaborn as sns
```

```python
with open("vetting_playlist.json", "r", encoding="utf-8") as f:
    data = json.load(f)

# Inspect the type of data
print(type(data))


tdf = pd.json_normalize(data)
tdf.info()

tdftracks=[]
for i in tdf['tracks.items']:
    tdftracks.append(i)
```

```python
ttracks=pd.DataFrame(tdftracks)
```
